 **Apriori Algorithm**
  
  The Apriori algorithm is a classic algorithm for association rule mining, based on the idea of frequent itemset generation. The algorithm starts by identifying the frequent itemsets and then generates the association rules based on these frequent itemsets.

```python
#installing apyori
!pip install apyori
```

```python
from apyori import apriori 
```

```python
#encoding
products = data['itemDescription'].unique()
dummy = pd.get_dummies(data['itemDescription'])
data.drop(['itemDescription'], inplace =True, axis=1)

data = data.join(dummy)

data.head()
```

```python
data1 = data.groupby(['Member_number', 'Date'])[products[:]].sum()
data1 = data1.reset_index()[products]

print("New Dimension", data1.shape)
data1.head()
```

```python
def productnames(x):
    for product in products:
        if x[product] >0:
            x[product] = product
    return x

data1 = data1.apply(productnames, axis=1)
data1.head()
```

```python
#Extracting the list of items bought per customer

x = data1.values
x = [sub[~(sub==0)].tolist() for sub in x if sub [sub != 0].tolist()]
transactions = x
transactions[0:10]
```

```python
rules = apriori(transactions, min_support = 0.00030, min_confidence = 0.05, min_lift = 3, max_length = 2, target = "rules")
association_results = list(rules)
print(association_results[0])
```

```python
for item in association_results:
    
    pair = item[0]
    items = [x for x in pair]
    
    print("Rule : ", items[0], " -> " + items[1])
    print("Support : ", str(item[1]))
    print("Confidence : ",str(item[2][0][2]))
    print("Lift : ", str(item[2][0][3]))
    
    print("--------------------------------") 
```

**FP Growth Algorithm**

The FP-growth algorithm is another popular algorithm for association rule mining. It is based on the idea of frequent pattern growth. The algorithm starts by identifying the frequent patterns and then generates the association rules based on these frequent patterns. It doesn't need  to generate candidates unlike apriori algorithm which makes it easier to use over large datasets.

```python
!pip install pyfpgrowth
```

```python
import pyfpgrowth
```

```python
transactions[0:10]
```

```python
# Identify frequent itemsets
patterns = pyfpgrowth.find_frequent_patterns(transactions, 10)
patterns
sorted_patterns = {key: value for key, value in sorted(patterns.items(), key=lambda item: item[1],reverse=True)}

sorted_patterns
```

```python
rules = pyfpgrowth.generate_association_rules(patterns, 0.02)
rules
```

```python
association_results = list(rules)
for item in association_results:
    print(item)
```
